```python
board = dict()
for i in range(1,10):
    board[i] = '_'
```

```python
def print_board():
    print(f"{board[1]}  |  {board[2]}  |  {board[3]}\t\t\t1,2,3\n")
    print(f"{board[4]}  |  {board[5]}  |  {board[6]}\t\t\t4,5,6\n")
    print(f"{board[7]}  |  {board[8]}  |  {board[9]}\t\t\t7,8,9\n")
```

```python
def is_full():
    draw = True
    for i in range(1,10):
        if board[i]=='_':
            draw = False
    if draw == True:
        return True
    else:
        return False
```

```python
def  minimax(board,depth,ismaximize,alpha,beta):
    global count
    c_mark = 'o'
    p_mark = 'x'
    if win(c_mark):
        count = count + 1
        return 10 - depth
    elif win(p_mark):
        count = count + 1
        return -10 + depth
    elif is_full():
        count = count + 1
        return 0
    if ismaximize:
        best_score = -100
        for i in range(1,10):
            if board[i]=='_':
                board[i]=c_mark
                score = minimax(board,depth+1,False,alpha,beta)
                if score>best_score:
                    best_score=score
                board[i]='_'
                alpha = max(alpha,score)
                if alpha>=beta:
                    break
        return best_score
    else:
        best_score = 100
        for i in range(1,10):
            if board[i]=='_':
                board[i]=p_mark
                score = minimax(board,depth+1,True,alpha,beta)
                if score<best_score:
                    best_score=score
                board[i]='_'
                beta = min(beta,score)
                if alpha>=beta:
                    break
        return best_score
```
